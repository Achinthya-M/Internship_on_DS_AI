```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("../data/customer_analytics.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
num_cols = df.select_dtypes(include=np.number).columns
cat_cols = df.select_dtypes(include="object").columns

df[num_cols] = df[num_cols].fillna(df[num_cols].median())
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['AnnualIncome'], kde=True)
plt.title("Annual Income Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Age', y='AnnualIncome', data=df)
plt.title("Age vs Annual Income")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Gender', y='AnnualIncome', data=df)
plt.title("Income by Gender")
plt.show()
```

```python
plt.figure(figsize=(8,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
df.to_csv("../data/customer_analytics_cleaned.csv", index=False)
```
