Machine Learning is a way of training computers to learn patterns from data rather than programming rules by hand.In traditional programming, we would provide the data and the set of rules to produce an output. In Machine Learning, we provide the data and the correct output, and the computer learns the rules or patterns between them automatically. This enables computers to make predictions or come to a decision on new data without being programmed for each scenario.

Supervised Learning
Supervised Learning is a machine learning technique that involves a dataset with labeled samples. This implies that every input sample has a corresponding correct output. The machine learning model is trained using these input-output pairs to make predictions on new samples.

Unsupervised Learning
Unsupervised Learning is applied when the data does not have labeled output values. The algorithm attempts to find hidden patterns or structures in the data without knowing the actual output values.

Reinforcement Learning
Reinforcement Learning is a type of learning method where an agent is allowed to interact with an environment and learn by receiving rewards or penalties for actions taken. The aim is to achieve maximum rewards in the long run.

| Learning Type          | Description                              | Example 1               | Example 2               |
|------------------------|------------------------------------------|-------------------------|-------------------------|
| Supervised Learning    | Data has labeled outputs                 | Spam Detection          | House Price Prediction  |
| Unsupervised Learning  | No labels, finds hidden patterns         | Customer Segmentation   | Market Basket Analysis  |
| Reinforcement Learning | Learns through rewards and penalties     | Self-Driving Cars       | Game Playing AI         |
